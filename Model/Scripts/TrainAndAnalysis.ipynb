## Model Analysis

```python
with open(Settings.log_save_folder + "train.json", "r") as file:
    log1 = json.load(file)
with open(Settings.log_save_folder + "trainAfter20.json", "r") as file:
    log2 = json.load(file)
with open(Settings.log_save_folder + "trainAfter40.json", "r") as file:
    log3 = json.load(file)
```

```python
log = dict()
for key in log1:
    log[key] = log1.get(key) + log2.get(key) + log3.get(key)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(log['train_loss'], '--', color=(1,0,0,0.8), label='BCE_LOSS (Train)')
plt.plot(log['valid_loss'], '--', color=(0,0,1,0.8), label='BCE_LOSS (Validation)')
plt.tight_layout()
plt.legend(loc="best")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(log['train_accu'], '--', color=(1,0,0,0.8), label='Accuracy (Train)')
plt.plot(log['valid_accu'], '--', color=(0,0,1,0.8), label='Accuracy (Validation)')
plt.tight_layout()
plt.legend(loc="best")
plt.show()
```

## Confusion Matrix

```python
from Evaluate import Evaluator
model = Model()
evaluator = Evaluator(model, "modelAfter40.h5")
```

```python
evaluator.record_matrix(test_loader, "confusion.json")
```

```python
import numpy as np
import pandas as pd
import seaborn as sn
```

```python
matrix = np.zeros((Settings.class_num, Settings.class_num))
with open(Settings.log_save_folder + 'confusion.json', 'r') as f:
    tuple_list = json.load(f)
```

```python
for point in tuple_list:
    i, j = point
    matrix[i][j] += 1
```

```python
for i, line in enumerate(matrix):
    matrix[i] /= sum(line)
matrix = pd.DataFrame(matrix, index = [evaluator.index_to_label.get(str(i+1)) for i in range(Settings.class_num)],
                  columns = [evaluator.index_to_label.get(str(i+1)) for i in range(Settings.class_num)],)
```

```python
plt.figure(figsize=(14, 10))
plt.title('Model')
sn.heatmap(matrix, annot=True,)
plt.tight_layout()
plt.show()
```
